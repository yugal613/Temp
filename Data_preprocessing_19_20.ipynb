# Batting Cleaning

```python
most_runs = pd.read_excel('C:\\Users\\BTC\\Desktop\\DPDM DATA\\most_runs.xlsx')
most_runs
```

```python
highstr_rate=pd.read_excel('C:\\Users\\BTC\\Desktop\\DPDM DATA\\highstr_rate.xlsx')
highstr_rate.head()
```

```python
most_runs.head()
```

```python
highstr_rate.head()
```

```python
combined_df = pd.concat([most_runs, highstr_rate])
combined_df
```

```python
unique_df = combined_df.drop_duplicates()
unique_df
```

```python
most_ducks=pd.read_excel('C:\\Users\\BTC\\Desktop\\DPDM DATA\\most_ducks.xlsx')
most_ducks
```

```python
most_ducks.shape
```

```python
combined = pd.concat([unique_df, most_ducks])
combined
```

```python
unique = combined.drop_duplicates()
unique
```

```python
most_ducks.columns
```

```python
ducks=most_ducks[['Player',0]]
ducks
```

```python
unique
```

```python
unique[0] = unique[0].replace('-', 0)
```

```python
unique['NO'] = unique['NO'].replace('-', 0)
```

```python
unique = unique.replace('-', 0)
```

```python
unique['HS'] = unique['HS'].astype(str)
```

```python
unique
```

```python
unique['highest_not_out'] = unique['HS'].apply(lambda x: 1 if '*' in x else 0)
```

```python
unique['HS'] = unique['HS'].str.replace('*', '', regex=False).astype(int)
```

```python
unique
```

```python
unique[['Player', 'State']] = unique['Player'].str.extract(r'(?P<Player>.+) \((?P<State>.+)\)')
```

```python
unique
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(unique['Runs'], bins=10, kde=True)
plt.title('Distribution of Runs')
plt.xlabel('Runs')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.heatmap(unique.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.xlabel('Columns')
plt.ylabel('Rows')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=unique[['Runs', 'Ave', 'BF', 'SR']])
plt.title('Box Plot for Runs, Batting Average, Balls Faced, and Strike Rate')
plt.show()
```

```python
sns.pairplot(unique[['Runs', 'Ave', 'BF', 'SR', 100, 50]], diag_kind='kde')
plt.title('Pair Plot for Selected Features')
plt.show()
```
